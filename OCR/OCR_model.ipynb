4. Set tesseract path

```python
import pytesseract
import cv2
import matplotlib.pyplot as plt

# Set Tesseract Path (Change This)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

***pytesseract***<br>
is a wrapper for the Tesseract-OCR engine, which is a popular open-source OCR engine

***cv2***<br>
is a computer vision library that provides a variety of functions for working with images

### OCR - Sample Text

```python
# Import the image file
image = cv2.imread("image/Sample Eng.jpg")
plt.imshow(image)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to the image
thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)[1]

# Recognize the text in the image
text = pytesseract.image_to_string(thresh)

# Print the text
print(text)
```
